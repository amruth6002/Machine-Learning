```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
#positive class
x1=np.random.normal(loc=5,scale=1,size=4)
y1=np.random.normal(loc=8,scale=1,size=4)
#negative class
x2=np.random.normal(loc=8,scale=1,size=4)
y2=np.random.normal(loc=5,scale=1,size=4)
```

```python
plt.scatter(x1,y1)
plt.scatter(x2,y2,color="red",marker='x')
plt.grid(True)
```

```python
def plot_d(theta,theta0,iter):
    plt.figure(figsize=(6,4))
    plt.scatter(x1,y1)
    plt.scatter(x2,y2)
    x_val=np.linspace(2,10,10)
    if(theta[1]!=0):
        y_val=-(theta[0]*x_val+theta0)/theta[1]
        plt.plot(x_val,y_val,label=f"{iter}")
    else:
        x_val=-theta0/theta[0]
        plt.axvline(x=x_val, label=f"{iter}")
    plt.legend()
```

```python
def percep(data,labels,k,d):
    theta=np.zeros(d)
    theta0=0
    for t in range(k):
        for i,xi in enumerate(data):
           if labels[i]*(np.dot(xi,theta)+theta0) <= 0 :
                theta+=labels[i]*xi
                theta0+=labels[i]
        plot_d(theta,theta0,t)
    return theta,theta0
```

```python
k=10
d=2
data1=np.vstack((x1,y1)).T
data2=np.vstack((x2,y2)).T
data=np.vstack((data1,data2))
labels=np.array([1]*len(data1)+[-1]*len(data2))
# print(labels)
best_theta,best_theta0=percep(data,labels,k,d)
```
